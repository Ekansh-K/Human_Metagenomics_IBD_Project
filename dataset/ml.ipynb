# ML model

- The problem we are handling right now is called as  single-label multi-class classification problem. In this type of problem, each instance is assigned exactly one label from a set of multiple possible classes. In your case, the classes are "brain", "heart", "lung", and "kidney".

SVM and Logistic Regression cannot handle Multi-Class classification Natively, and hence we use One-vs-Rest (OvR) or One-vs-One (OvO) strategies to handle multi-class classification.

Several machine learning algorithms can handle multi-class classification natively, meaning they are designed to work directly with multiple classes without needing additional strategies like One-vs-Rest (OvR) or One-vs-One (OvO). Some examples of algorithms that support multi-class classification natively include:
- decision tree classifier >
- random forest classifier >
- Gradieent Boosting Classifier 
- Naive Bayes: GaussianNB
- K-Nearest Neighbors (KNN) >
- Neural Networks like Multi-Layer Perceptron






Regression models are not suitable for single-label multi-class classification problems. Regression models are designed to predict continuous numerical values, whereas single-label multi-class classification involves predicting one label from a set of discrete classes.

```python
df = pd.read_csv("lil_data.csv")
display(df)
```

Data Splitting:

```python
Input_Features = [col for col in df.columns if col != 'disease']
X = df[Input_Features]
Y = df['disease']
# Splitting the data into training and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

# 1. Multi-Class Classifiation using SVM

for this, we will be using the One-vs-Rest strategy. In this strategy, we fit one classifier per class. For each classifier, we assign the class as the positive class and all other classes as the negative class. We then predict the class with the highest confidence score.

```python
from sklearn.multiclass import OneVsRestClassifier
basemodel = SVC()
model = OneVsRestClassifier(basemodel)
model.fit(X_train, Y_train)
Y_predORSVC = model.predict(X_test)
print('Accuracy: ', accuracy_score(Y_test, Y_predORSVC))
print('Confusion Matrix: ', confusion_matrix(Y_test, Y_predORSVC))
print('Classification Report: ', classification_report(Y_test, Y_predORSVC))
```

# 2. Logistic Regression
Here also we will be using One-vs-Rest strategy.
>One might ask: "why are we using regression here when we are dealing with classification problem?" The answer is that logistic regression is a classification algorithm. The name "regression" in logistic regression is a misnomer. Logistic regression is a classification algorithm that predicts the probability of an instance belonging to a particular class.

```python
modelLR = LogisticRegression()
multimodelLR = MultiOutputRegressor(modelLR)
multimodelLR.fit(X_train, Y_train.values.reshape(-1, 1))
Y_predLR = multimodelLR.predict(X_test).reshape(-1)
print("Accuracy: ", accuracy_score(Y_test, Y_predLR))
print("Confusion Matrix: \n", confusion_matrix(Y_test, Y_predLR))
print("Classification Report: \n", classification_report(Y_test, Y_predLR))
```
